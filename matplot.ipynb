```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x= np.arange(0,10,1)
y=np.arange (0,10,1)
```

```python
# sourcery skip: my-custom-rule
plt.plot(x,y, label = 'linear')
plt.plot(x, x**2, label = 'exp x')
plt.plot(y, y**3, label = 'exp y')

plt.title('First Graph', fontdict = { 'fontsize': 12 })
 
# Adding labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks(x)
plt.yticks(y)

# Show the  graph legend
plt.legend()

plt.savefig('mygraph.png', dpi=300)

# Removes the auto-heading
plt.show() 

```

```python
labels =['A','B','C']
values = [1,2,3]

bars = plt.bar(labels,values)

# Shading bars using a for loop
partterns = ['/','*','.']
for bar in bars:
    bar.set_hatch(partterns.pop(0))
    
    # An alternative
# bars[0].set_hatch('/')
# bars[1].set_hatch('*')
# bars[2].set_hatch('.')

plt.figure(figsize=(6,4))
plt.savefig('Bar 1')
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins = bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number Of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.savefig('Histogram', dpi = 300)

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']== 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['y', 'g']
plt.pie([left, right], labels = labels, colors = colors, autopct ='%.2f %%')

plt.title('Foot Preferrence Of FIFA Players')

plt.savefig('Pie 1')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

plt.style.use('ggplot')
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >=150) & (fifa.Weight <175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

explode = (.4,.4,0,0,.4)

weights = [light, light_medium, medium, medium_heavy, heavy]
lables = ['Under 125','125-150','150-175','175-200','Over 200']

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels = lables , autopct='%.2f %%',explode= explode, pctdistance= 0.8)

plt.savefig('Pie 2')

plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=(5,8))
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['barcelona', 'madrid', 'revs']
boxes = plt.boxplot([barcelona, madrid, revs], labels= labels, patch_artist=True)

plt.savefig('Candlesticks')

for box in boxes['boxes']:
    box.set(color='#428ef4', linewidth=2)
    box.set(facecolor='#e0e0e0')
    
plt.title('Professional Soccer Teams Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```
