```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x= np.arange(0,10,1)
y=np.arange (0,10,1)

```

```python
plt.plot(x,y, label = 'linear')
plt.plot(x, x**2, label = 'exp x')
plt.plot(y, y**3, label = 'exp y')

plt.title('First Graph', fontdict = { 'fontsize': 12 })
 
# Adding labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks(x)
plt.yticks(y)

# Show the  graph legend
plt.legend() 

plt.savefig('mygraph.png', dpi=300)

# Removes the auto-heading
plt.show() 

```
