```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x= np.arange(0,10,1)
y=np.arange (0,10,1)
```

```python
plt.plot(x,y, label = 'linear')
plt.plot(x, x**2, label = 'exp x')
plt.plot(y, y**3, label = 'exp y')

plt.title('First Graph', fontdict = { 'fontsize': 12 })
 
# Adding labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks(x)
plt.yticks(y)

# Show the  graph legend
plt.legend() 

plt.savefig('mygraph.png', dpi=300)

# Removes the auto-heading
plt.show() 

```

```python
labels =['A','B','C']
values = [1,2,3]

bars = plt.bar(labels,values)

# Shading bars using a for loop
partterns = ['/','*','.']
for bar in bars:
    bar.set_hatch(partterns.pop(0))
    
    # An alternative
# bars[0].set_hatch('/')
# bars[1].set_hatch('*')
# bars[2].set_hatch('.')

plt.figure(figsize=(6,4))

plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins = bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number Of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']== 'Right'].count()[0]

labels = ['Left', 'Right']
colors = ['y', 'g']
plt.pie([left, right], labels = labels, colors = colors, autopct ='%.2f %%')

plt.title('Foot Preferrence Of FIFA Players')

plt.show()
```
